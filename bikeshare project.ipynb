```python
import pandas as pd
import numpy as np
import time

CITY_DATA = { 'chicago': 'D:/hopfully a career/fwd/PYTHON FWD PROJECTS/chicago.csv',
              'new york city': 'D:/hopfully a career/fwd/PYTHON FWD PROJECTS/new_york_city.csv',
              'washington': 'D:/hopfully a career/fwd/PYTHON FWD PROJECTS/washington.csv' }
months = ['january', 'february', 'march', 'april', 'may', 'june']
days = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday','saturday', 'sunday']






def get_filters():
    """
    Asks user to specify a city, month, and day to analyze.

    Returns:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    """
    print('Hello! Let\'s explore some US bikeshare data!')
    while True:
        city = input('Which city would you like to explore? \'chicago\', \'new york city\', or \'washington\'?').lower()
        if city in CITY_DATA.keys():
            break
        else:
            print('\nInvalid input. Try again.\n')
            continue
    while True:
        month = input('Which month would you like to explore? january,february,etc').lower()
        if month == 'all':
            print('\nokay lets explore data for all months.')
            break
        elif month in months:
            print('\nLets get the data for {}.'.format(month.title()))
            break
        else:
            print('\nInvalid input.  Try again.\n')
            continue
    while True:
        day = input('Which day would you like to explore?').lower()
        if day == 'all':
            print('\nokay lets explore data for all days')
            break
        elif day in days:
            print('\nokay lets explore data for {}'. format(day.title()))
            break
        else:
            print('\nInvalid input. Try again.\n')
            continue

    print('-'*40)
    return city, month, day


def load_data(city, month, day):
    """
    Loads data for the specified city and filters by month and day if applicable.

    Args:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    Returns:
        df - Pandas DataFrame containing city data filtered by month and day
    """
    df = pd.read_csv(CITY_DATA[city])
    df['Start Time'] = pd.to_datetime(df['Start Time'])
    df['Month'] = df['Start Time'].dt.month_name()
    df['Day'] = df['Start Time'].dt.day_name()
    df['Hour'] = df['Start Time'].dt.hour
    
    #filtering by months
    if month == 'all' and day != 'all':
        months_df = df[(df['Day']==day.title())]
        print('-'*40)
        return months_df
    #Filtering by day 
    elif day == 'all' and month != 'all':
        days_df = df[(df['Month']==month.title())]
        print('-'*40)
        return days_df

    #Filtering by month and day
    elif month != 'all' and day != 'all':
        df_specific = df[(df['Month']==month.title()) &(df['Day']==day.title())]
        print('-'*40)
        return df_specific
    else:  
        print('-'*40)
        return df


def time_stats(df):
    """Displays statistics on the most frequent times of travel."""

    print('\nCalculating The Most Frequent Times of Travel...\n')
    start_time = time.time()

    # TO DO: display the most common month
    common_month = df['Month'].mode()[0]
    print('Most common month in this data set is {}.'.format(common_month))

    # TO DO: display the most common day of week
    common_day = df['Day'].mode()[0]
    print('Most common day in this data set is {}.'.format(common_day))

    # TO DO: display the most common start hour
    common_hour = df['Hour'].value_counts().idxmax()
    print('Most common hour in this data set is {}.'.format(common_hour))

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def station_stats(df):
    """Displays statistics on the most popular stations and trip."""

    print('\nCalculating The Most Popular Stations and Trip...\n')
    start_time = time.time()

    # TO DO: display most commonly used start station
    common_sstation = df['Start Station'].mode()[0]
    print('Most common starting station in this data set is {}.'.format(common_sstation))
    
    

    # TO DO: display most commonly used end station
    common_estation = df['End Station'].mode()[0]
    print('Most common ending station in this data set is {}.'.format(common_estation))

    # TO DO: display most frequent combination of start station and end station trip
    grouping = df.groupby(['Start Station','End Station'])
    pop_Station = grouping.size().sort_values(ascending = False).head(1)
    print('Most frequent combination of Start Station and End Station trip:\n', pop_Station)
    
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def trip_duration_stats(df):
    """Displays statistics on the total and average trip duration."""

    print('\nCalculating Trip Duration...\n')
    start_time = time.time()

    # TO DO: display total travel time
    t_sum = df['Trip Duration'].sum()
    print('The total time of all trips in this data set is {:,.0f}seconds.'.format(t_sum))
    print('That is {:,.2f} days.'.format(t_sum/60/60/24))

    # TO DO: display mean travel time
    t_mean = df['Trip Duration'].mean()
    print('The average time of all trips in this data is {:,.2f} seconds.'.format(t_mean))
    print('That is {:,.2f} minutes.'.format(t_mean/60))
    
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def user_stats(df,city):
    """Displays statistics on bikeshare users."""

    print('\nCalculating User Stats...\n')
    start_time = time.time()

    # TO DO: Display counts of user types
    
    u = df['User Type'].value_counts()
    print('The following shows the count of each user type:\n', u)

    # TO DO: Display earliest, most recent, and most common year of birth
    if city != 'washington':
        print('Birth Year Stats:')
        most_common_year = df['Birth Year'].mode()[0]
        print('Most Common Year:',most_common_year)
        most_recent_year = df['Birth Year'].max()
        print('Most Recent Year:',most_recent_year)
        earliest_year = df['Birth Year'].min()
        print('Earliest Year:',earliest_year)
    
    # TO DO: Display counts of gender
    
        g = df['Gender'].value_counts()
        print('The following shows the count of male and female users:\n', g)

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
#view row data to user
def show_data(df):
    start_loc = 0
    while True:
        view_data = input("Would you like to view 5 rows of individual trip data? Enter yes or no?").lower()
        if view_data == 'yes':
            print(df.iloc[start_loc:start_loc+6])
            start_loc += 5
        elif view_data == 'no':
            break
        else:
            print('Invalid input. Please try again')
    

def main():
    while True:
        city, month, day = get_filters()
        df = load_data(city, month, day)

        time_stats(df)
        station_stats(df)
        trip_duration_stats(df)
        user_stats(df,city)
        show_data(df)

        restart = input('\nWould you like to restart? Enter yes or no.\n')
        if restart.lower() != 'yes':
            break


if __name__ == "__main__":
	main()
```
